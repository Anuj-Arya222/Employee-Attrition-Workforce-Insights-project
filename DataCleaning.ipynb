```python
df=pd.read_csv('employee_churn_200k.csv')
```

```python
print(df.info())
print(df.describe())
print(df.columns)
```

```python
df.head()
```

```python
# fix target value
df['Attrition']=df['Attrition'].map({'Yes':1,'No':0})
```

```python
df['Attrition'].head(10)
```

```python
# overall attrition Rate
df['Attrition'].mean()
```

```python
df.groupby('Overtime')['Attrition'].mean()*100
# Without overtime → only 21% attrition
# With overtime → 51% attrition 😳

# Key Insight:
# Overtime is a major driver of attrition
# Employees working overtime are more than twice as likely to leave
# Indicates burnout and poor work-life balance

# Reduce excessive overtime
# Improve workload distribution
# Promote work-life balance policies
```

```python
df.head()
```

```python
df.groupby('JobSatisfaction')['Attrition'].mean()*100

# Employees with:
# Low satisfaction (1–2) → ~46% attrition 🚨
# High satisfaction (3–4) → ~26% attrition
# Low satisfaction employees are ~1.8x more likely to leave

#  Notice:
# Satisfaction 1 & 2 are almost same
# Satisfaction 3 & 4 are almost same

#  This means:
# Employees don’t leave gradually — they leave when they are unhappy
# Job satisfaction is a major factor influencing attrition
# Employees with low satisfaction are almost twice as likely to leave
# Improving satisfaction from low → medium has high impact

# BUSINESS RECOMMENDATION
# Focus on low satisfaction employees (1 & 2)
# Conduct employee feedback surveys
# Improve work environment & management
```

```python
df.groupby(['Overtime','JobSatisfaction'])['Attrition'].mean()
# No Overtime:
# Low satisfaction → ~31%
# High satisfaction → ~11%
# With Overtime:
# Low satisfaction → ~61% 
# High satisfaction → ~41% 
# Even happy employees leave if they work overtime
#  Worst Case:

#  Overtime + Low Satisfaction → 61% attrition 

# Shocking Insight:

# Even with high satisfaction (4):
# No overtime → 11%
# With overtime → 41% 

# Means:
# Overtime destroys retention
```
